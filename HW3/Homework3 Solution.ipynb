## 6 Fun with Intervals

```python
import random

N = [3, 4, 5, 6, 7]

for i in N:
    random_numbers = [random.choice([-1, 1]) for num in range(i)]
    print(random_numbers)
```

For random chosen number, when N=5, it cannot be separated by two intervals, so answer is 5, c.

```python
for i in range(2, 7):
    print(2**i)
```

```python
from scipy.special import comb 

def alt_a(N):
    return comb(N+1, 4)

def alt_b(N):
    return comb(N+1, 2) + 1

def alt_c(N):
    return comb(N+1, 4) + comb(N+1, 2) + 1

def alt_d(N):
    return comb(N+1, 4) + comb(N+1, 3) + comb(N+1, 2) + comb(N+1, 1) + 1
```

```python
print("\nA")
for i in range(1, 7):
    print(i, alt_a(i))

print("\nB")    
for i in range(1, 7):
    print(i, alt_b(i))

print("\nC")    
for i in range(1, 7):
    print(i, alt_c(i))
    
print("\nD")    
for i in range(1, 7):
    print(i, alt_d(i))
```

So the answer of 6 is c.

The answer of 7 is 2M+1, since 2*2+1 = 5.

## 10 Non-Convex Sets: Concentric Circles

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Define the center coordinates
center = (0, 0)

# Define the radii of the circles
radii = [1, 2]

# Draw the circles
for radius in radii:
    circle = plt.Circle(center, radius, edgecolor='black', fill=False)
    ax.add_patch(circle)

# Set the aspect ratio to equal
ax.set_aspect('equal')

# Set the x and y limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Concentric Circles')

# Display the plot
plt.show()
```

The answer is b, it is like interval case when all points are in a line.
